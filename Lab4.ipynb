##<b>Manual Calculation For Dot Product Of Vector K and J:

```
k=np.array([1,-1,2,-2])
j=np.array([3,-3,-4,4])

R1=k[0]*j[0]
R2=k[1]*j[1]
R3=k[2]*j[2]
R4=k[3]*j[3]

dp=R1+R2+R3+R4
print(dp)
```
